```python
def filter_prime_lines(input_file, output_file):
    prefix = "***is prime  "
    with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
        for line in infile:
            if line.startswith(prefix):
                number = line[len(prefix):].strip()
                outfile.write(number + '\n')

if __name__ == "__main__":
    input_file = 'filtered_output.txt'  # Replace with your input file name
    output_file = 'filtered_output.csv'  # Replace with your desired output file name
    filter_prime_lines(input_file, output_file)
    print(f"Filtered lines have been written to {output_file}")
```
