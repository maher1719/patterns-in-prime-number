```python
def parse_input(input_data):
    # Split the input into lines
    lines = input_data.strip().split("\n")
    
    # Read the number of test cases
    t = int(lines[0])
    index = 1  # Track the current line
    
    # Initialize a list to hold all test cases
    test_cases = []
    
    # Loop through the test cases
    for _ in range(t):
        # Read n and m
        n, m = map(int, lines[index].split())
        index += 1
        
        # Read the array a
        a = list(map(int, lines[index].split()))
        index += 1
        
        # Read the array b
        b = list(map(int, lines[index].split()))
        index += 1
        
        # Append the test case data as a tuple or dictionary
        test_cases.append({
            "n": n,
            "m": m,
            "a": a,
            "b": b
        })
    
    return test_cases

# Example usage
input_data = """9
2 1
4 5
9
2 1
3 6
9
4 2
1 2 2 2
3 4
4 2
1 1 3 3
3 5
4 2
1 2 3 4
3 5
5 5
1 2 3 4 5
5 4 3 2 1
4 2
1 1 1 1
1 1
4 4
1 1 1 1
1 1 1 2
1 1
1
1000000000"""

test_cases = parse_input(input_data)

# Output the parsed test cases
for i, case in enumerate(test_cases, start=1):
    print(f"Test Case {i}:")
    print(f"  n = {case['n']}, m = {case['m']}")
    print(f"  a = {case['a']}")
    print(f"  b = {case['b']}")
    print()
```
