```python
import math
import csv

# Cache for prime numbers
prime_cache = {}

# Function to check if a number is prime
def is_prime(num):
    if len(prime_cache) > math.pow(10, 7):
        prime_cache.clear()
    if len(prime_cache) >= math.pow(2, 6):
        prime_cache.clear()
    if num in prime_cache:
        return prime_cache[num]
    if num <= 1:
        return False
    if num <= 3:
        return True
    if num % 2 == 0 or num % 3 == 0:
        return False

    i = 5
    while i * i <= num:
        if num % i == 0 or num % (i + 2) == 0:
            prime_cache[num] = False
            return False
        i += 6

    prime_cache[num] = True
    return True

# Function to generate primes and twin primes
def addition(number_to_generate_primes):
    quotient = math.floor(number_to_generate_primes)
    second_mult = 0
    third_mult = 0
    primes = []
    multiple_minus_one = 0
    twins_number = 0

    for index in range(2, number_to_generate_primes + 1):
        possible_twin1 = None
        possible_twin2 = None
        prime_plus_one = index * quotient + 1
        prime_minus_one = index * quotient - 1

        if is_prime(prime_plus_one):
            second_mult += 1
            possible_twin1 = prime_plus_one
            primes.append(prime_plus_one)

        if is_prime(prime_minus_one):
            multiple_minus_one += 1
            possible_twin2 = prime_minus_one
            primes.append(prime_minus_one)

        if possible_twin1 and possible_twin2 and abs(possible_twin1 - possible_twin2) == 2:
            twins_number += 1

    return second_mult, third_mult, primes, multiple_minus_one, twins_number

# Function to calculate prime numbers within a range
def calculate(min_interval, end_interval):
    sqrt2 = math.sqrt(2)
    unique_primes = []
    labels = []
    prime_count = []
    prime_count_minus_one_equation = []
    prime_twins_count = []
    ln_func = []
    ln_func2 = []
    ln_func3 = []
    ln_func4 = []
    ln = []

    primes_count_plus = 0
    primes_count_minus = 0

    for number in range(min_interval, end_interval + 1):
        result = addition(number)
        prime_check = is_prime(number)
        labels.append(number)
        log_number = math.log(number)- sqrt2
        prime_count.append(result[0])
        primes_count_plus += result[0]
        primes_count_minus += result[3]
        prime_count_minus_one_equation.append(result[3])
        prime_twins_count.append(result[4])
        mathLogNumber=round(number / log_number)
        ln_func.append(mathLogNumber)
        ln_func2.append(mathLogNumber * sqrt2)
        ln_func3.append(mathLogNumber / sqrt2)
        ln_func4.append(mathLogNumber * 2 * sqrt2)
        
        ln.append(log_number)

        for prime in result[2]:
            unique_primes.append(prime)

    unique_primes_set = sorted(set(unique_primes))
    twin_primes = [unique_primes_set[i] for i in range(1, len(unique_primes_set))
                   if unique_primes_set[i] - unique_primes_set[i - 1] == 2]

    prime_results = {
        "unique_primes_count": len(unique_primes_set),
        "twin_primes_count": len(twin_primes),
        "primes_count_plus": primes_count_plus,
        "primes_count_minus": primes_count_minus
    }

    # CSV output (for saving results)
    with open('prime_data.csv', 'w', newline='') as csvfile:
        fieldnames = ['Number', 'Primes +1', 'Primes -1', 'Twins', 'Is Prime']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for i, label in enumerate(labels):
            writer.writerow({
                'Number': label,
                'Primes +1': prime_count[i],
                'Primes -1': prime_count_minus_one_equation[i],
                'Twins': prime_twins_count[i],
                'Is Prime': is_prime(label)
            })

    return prime_results, twin_primes

# Example of using the functions
if __name__ == "__main__":
    min_interval = int(input("Enter the minimum interval: "))
    end_interval = int(input("Enter the maximum interval: "))
    
    result, twin_primes = calculate(min_interval, end_interval)
    
    print(f"Unique primes count: {result['unique_primes_count']}")
    print(f"Twin primes count: {result['twin_primes_count']}")
    print(f"Primes +1: {result['primes_count_plus']}")
    print(f"Primes -1: {result['primes_count_minus']}")
    print(f"Twin primes found: {', '.join(map(str, twin_primes))}")
```
